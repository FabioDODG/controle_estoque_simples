```python
## controle de estoque automatizado.
print("""
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Cada tipo de bebida deve conter ao menos 400 unidades.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Cada tipo de comida deve conter ao menos 300 unidades.  
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
""")

lista_bebidas= {'Coca Cola': 250, 'Guarana Antartica': 310, 'Sprite': 400, 'Dell Vale Uva': 285, 'Dell Vale Pessego': 116 ,'Guaracamp': 900}
## contabiliza a quantidade de cada bebida no estoque.

lista_comidas= {'Pastel de frango': 110, 'Pastel de carne': 250, 'Pastel de queio': 77, 'Coxinha': 190, 'Quibe': 80}
## contabiliza a quantidade de cada comida no estoque.

escolha_de_tipo= input("""
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'Bebida' para estoque de bebidas.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'Comida' para estoque de comidas.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
""").capitalize()

if escolha_de_tipo == "Bebida":
    print("""
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'C' para Coca Cola.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'GA' para Guarana Antartica.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'S' para Sprite.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'DP' para Dell Vale de Pessego.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'DU' para Dell Vale de Uva.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'G' para Guaracamp.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
 """)

    bebida= input('Digite sua escolha: ').upper()
    if bebida == "C":
        print('{}'.format(lista_bebidas['Coca Cola']))
        coca= 400 - lista_bebidas['Coca Cola']
        print("Deve ser comprados {} de Coca Cola.".format(coca))
    elif bebida =="GA":
        print('{}'.format(lista_bebidas['Guarana Antartica']))
        guarana= 400 - lista_bebidas['Guarana Antartica']
        print("Deve ser comprados {} de Guarana.".format(guarana))
    elif bebida =="S":
        print('{}'.format(lista_bebidas['Sprite']))  
        sprite= 400 - lista_bebidas['Sprite'] 
        print("Deve ser comprados {} de Sprite.".format(sprite))
    elif bebida =="DP":
        print('{}'.format(lista_bebidas['Dell Vale Pessego']))
        dell_pessego= 400 - lista_bebidas['Dell Vale Pessego']
        print("Deve ser comprados {} de Dell Vale de Pessego.".format(dell_pessego))
    elif bebida =="DU":
        print('{}'.format(lista_bebidas['Dell Vale Uva']))
        dell_uva= 400 - lista_bebidas['Dell Vale Uva']
        print("Deve ser comprados {} de Dell Vale de Uva.".format(dell_uva))
    elif bebida =="G":
        print('{}'.format(lista_bebidas['Guaracamp']))
        guaracamp= 400 - lista_bebidas['Guaracamp']
        print("Deve ser comprados {} de Guarana.".format(guaracamp))
    else:
        print('Digite novamente.')

if escolha_de_tipo == "Comida":
    print("""
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'PF' para Pastel de Frango.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'PC' para Pastel de Carne.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'PQ' para Pastel de Queijo.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'C' para Coxinha.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
Digite 'QI' para Quibe.
=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
""")
    comida= input('Digite sua escolha: ').upper()
    if comida == "PF":
        print('{}'.format(lista_comidas['Pastel de frango']))
        pf= 300 - lista_comidas['Pastel de frango']
        print("Deve ser comprados {} de Pastel de Frango.".format(pf))
    elif comida =="PC":
        print('{}'.format(lista_comidas['Pastel de carne']))
        pc= 300 - lista_comidas['Pastel de carne']
        print("Deve ser comprados {} de Pastel de Carne.".format(pc))
    elif comida =="PQ":
        print('{}'.format(lista_comidas['Pastel de queio']))  
        pq= 400 - lista_comidas['Pastel de queio'] 
        print("Deve ser comprados {} de Pastel de Queijo.".format(pq))
    elif comida =="C":
        print('{}'.format(lista_comidas['Coxinha']))
        coxinha= 300 - lista_comidas['Coxinha']
        print("Deve ser comprados {} de Coxinha.".format(coxinha))
    elif comida =="QI":
        print('{}'.format(lista_comidas['Quibe']))
        quibe= 300 - lista_comidas['Quibe']
        print("Deve ser comprados {} Quibe.".format(quibe))
    else:
        print('Digite novamente.')


```
